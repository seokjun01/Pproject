```python
import pandas as pd
```

```python
df = pd.read_excel('신흥반점.xlsx')
df
```

```python
df = df.loc[:, ["매장명", "닉네임", "review", "label"]]
df
```

```python
df.isnull().sum()
```

```python
df['review'].nunique() # 리뷰 변수에 있는 중복내용을 제외한 리뷰, 중복이 몇개나 있는지 확인
```

```python
user_input = input('검색 : ')

# 검색어가 포함된 리뷰 필터링
search_df = df[df['review'].str.contains(user_input)]

if len(search_df) == 0:
    print("검색 결과가 없습니다.")
else:
    for idx, 매장명, 닉네임, review, label in search_df.itertuples():
        # label 값을 "긍정" 또는 "부정"으로 변환
        sentiment = "긍정" if label == 1 else "부정"
        
        print("[{}] {}님의 {}리뷰 \n리뷰내용: {}".format(
            매장명, 닉네임, sentiment, review))
```
